# Notebook 01: Data Ingestion and Validation
This notebook loads the raw source CSV files, validates key columns, and inspects the initial data shape and quality.

```python
import pandas as pd
from pathlib import Path

ROOT_DIR = Path.cwd()
DATA_RAW = ROOT_DIR / "data" / "raw"

files = {
    "fund_master": DATA_RAW / "01_fund_master.csv",
    "nav_history": DATA_RAW / "02_nav_history.csv",
    "aum_by_fund_house": DATA_RAW / "03_aum_by_fund_house.csv",
    "monthly_sip_inflows": DATA_RAW / "04_monthly_sip_inflows.csv",
    "category_inflows": DATA_RAW / "05_category_inflows.csv",
    "industry_folio_count": DATA_RAW / "06_industry_folio_count.csv",
    "performance": DATA_RAW / "07_scheme_performance.csv",
    "investor_transactions": DATA_RAW / "08_investor_transactions.csv",
    "portfolio_holdings": DATA_RAW / "09_portfolio_holdings.csv",
    "benchmark_indices": DATA_RAW / "10_benchmark_indices.csv",
}

for name, path in files.items():
    print(f"{name}: {path.exists()} - {path.stat().st_size if path.exists() else 'missing'} bytes")
```

```python
# Load key raw tables and inspect sample rows
fund_master = pd.read_csv(files["fund_master"])
nav_history = pd.read_csv(files["nav_history"])
performance = pd.read_csv(files["performance"])
investor_transactions = pd.read_csv(files["investor_transactions"])

print("fund_master", fund_master.shape)
print("nav_history", nav_history.shape)
print("performance", performance.shape)
print("investor_transactions", investor_transactions.shape)

fund_master.head()
```

```python
# Basic validation checks for missing values and duplicates
for df_name, df in [
    ("fund_master", fund_master),
    ("nav_history", nav_history),
    ("performance", performance),
    ("investor_transactions", investor_transactions),
]:
    missing = df.isna().sum().sort_values(ascending=False).head(10)
    dup_count = df.duplicated().sum()
    print(f"{df_name}: duplicates={dup_count}")
    print(missing)
    print("---")
```
